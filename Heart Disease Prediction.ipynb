## Data Preprocessing

```python
dataset = pd.get_dummies(df,columns=['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
c_to_s = ['age','trestbps','chol','thalach','oldpeak']
dataset[c_to_s] = sc.fit_transform(dataset[c_to_s])
```

```python
dataset.head()
```

```python
y = dataset['target']
X = dataset.drop('target',axis='columns')
```

```python
from sklearn.model_selection import cross_val_score
knn_score = []
for k in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn,X,y,cv=10)
    knn_score.append(score.mean())
```

```python
plt.figure(figsize=(16,10))
plt.plot([k for k in range(1,21)],knn_score,color='red')
for i in range(1,21):
    plt.text(i,knn_score[i-1],(i,knn_score[i-1]))
plt.xticks([i for i in range(1,21)])
plt.xlabel("Number of Neighbors")
plt.ylabel("Scores")
plt.title("K Neighbors Classifier",size=30)
```
